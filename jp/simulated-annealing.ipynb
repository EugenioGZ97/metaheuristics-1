```python
import math
# funcion objetivo (Minimizar, siendo menor a 0)
# MAX 1/(x+0.1)
def obj_fun_1(x):
    return 1/(x+0.1)
# MIN 0.05x^3-1x^2+5
def obj_fun_2(x):
    return 0.05*x**3-1*x**2+5
# MIN (x/10)^5-5(x/10)^4+5(x/10)^3+5(x/10)^2-6(x/10)-1
def obj_fun_3(x):
    return (x/10)**5-5*(x/10)**4+5*(x/10)**3+5*(x/10)**2-6*(x/10)-1
def move_probability(objective, obj_func, x_i, x_new, temp):
    of_1 = obj_func(x_i)
    of_2 = obj_func(x_new)
    prob = min(1 / (math.e**((of_1-of_2)/temp)), 1)
    options = {
        'min': min(1 / (math.e**((of_2-of_1)/temp)), 1),
        'max': min(1 / (math.e**((of_1-of_2)/temp)), 1)
    }
    return options[objective]
```

```python
objective = 'min'
fun = obj_fun_2
x_i = 0
x_new = 1
temperature = 0.1


prob = move_probability(objective, fun, x_i, x_new, temperature)

f_i = fun(x_i)
f_n = fun(x_new)
if objective == 'min':
    case = 'mejor' if f_n < f_i else 'peor'
elif objective == 'max':
    case = 'mejor' if f_n > f_i else 'peor'


print(f'x actual = {x_i}')
print(f'x nuevo  = {x_new}')
print(f'el objetivo es {objective}imizar')
print(f'obj actual = {f_i}')
print(f'obj nuevo  = {f_n}')
print(f'El valor nuevo es {case} que el anterior')
print(f'nueva temperatura = {temperature}')
print(f'probability = {prob*100}%')
```

```python
# next steps
step = 5.0
alpha = 0.99 # temperature decrease
x_i = x_new
x_new = x_i + step
temperature = alpha * temperature

prob = move_probability(objective, fun, x_i, x_new, temperature)

f_i = fun(x_i)
f_n = fun(x_new)
if objective == 'min':
    case = 'mejor' if f_n < f_i else 'peor'
elif objective == 'max':
    case = 'mejor' if f_n > f_i else 'peor'


print(f'x actual = {x_i}')
print(f'x nuevo  = {x_new}')
print(f'el objetivo es {objective}imizar')
print(f'obj actual = {f_i}')
print(f'obj nuevo  = {f_n}')
print(f'El valor nuevo es {case} que el anterior')
print(f'nueva temperatura = {temperature}')
print(f'probability = {prob*100}%')
```
